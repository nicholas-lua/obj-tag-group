## 1. Installation and Setup

```python
from obj_tag_group import Tag_Group
```

```python
reg = Tag_Group()
```

## 2. Core Registry Operations
Let's initialize a tracking registry and insert a mix of standard and unhashable types (like raw lists and dictionaries) that standard sets cannot natively support.

```python
# Creating and tagging objects
test_1 = {1: "Apple", 2: ("Lemon", "Mango")}
test_2 = ["Mango", "Pineapple", "Cabbage"]
test_3 = "Lettuce"
test_4 = ("beef", "ham")

reg.tag_object(test_1, tags = ["dict", "fruit"])
reg.tag_object(test_2, tags = ["list", "fruit", "vegetable"])
reg.tag_object(test_3, tags = ["string", "vegetable"])
reg.tag_object(test_4, tags = ["tuple", "meat"])

```

```python
# Filtering Objects by Tags
# Example: Find all objects with both "fruit" and "vegetable")
results = reg.filter_tags(pattern=r"vegetable|fruit")
print(results)
```

```python
# 1. Saving registry to virtual disk
reg.save_to_file("demo_registry.pkl")
print("Successfully saved to disk!")

# 2. Load data from the file to a new registry
loaded_reg = Tag_Group.load_from_file("demo_registry.pkl")

# 3. Query the restored registry to prove the data is intact
restored_results = loaded_reg.filter_tags(pattern=r"string")
print("\nRestored Query Results:")
print(restored_results)
```
